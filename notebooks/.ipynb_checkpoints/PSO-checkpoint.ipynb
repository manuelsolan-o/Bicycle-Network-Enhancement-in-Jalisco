```python
import random
import numpy as np

# Define PSO parameters
num_particles = 50
num_dimensions = 300
max_iterations = 100
c1 = 2.0  # Cognitive coefficient
c2 = 2.0  # Social coefficient
w = 0.5  # Inertia weight

# Example demand data
demanda_estaciones = [random.randint(1, 15) for _ in range(num_dimensions)]

# Initialize particles
def initialize_particles(num_particles, num_dimensions):
    particles = []
    for _ in range(num_particles):
        particle = np.array(generar_individuos(num_dimensions, 5.5))
        while np.sum(particle) != limite_bicicletas:
            particle = np.array(generar_individuos(num_dimensions, 5.5))
        velocity = np.zeros(num_dimensions)
        personal_best_position = particle.copy()
        personal_best_fitness = evaluate_fitness(particle)
        particles.append({
            'position': particle,
            'velocity': velocity,
            'personal_best_position': personal_best_position,
            'personal_best_fitness': personal_best_fitness
        })
    return particles

# Evaluate fitness
def evaluate_fitness(position):
    exceso_bicicletas = np.sum(position) - limite_bicicletas
    fitness = np.sum(np.abs(np.array(position) - np.array(demanda_estaciones)))
    if exceso_bicicletas > 0:
        fitness += exceso_bicicletas  # Penalize the excess of bicycles
    return fitness

# PSO algorithm
particles = initialize_particles(num_particles, num_dimensions)
global_best_position = None
global_best_fitness = float('inf')

for _ in range(max_iterations):
    for particle in particles:
        r1, r2 = random.random(), random.random()
        cognitive_component = c1 * r1 * (particle['personal_best_position'] - particle['position'])
        social_component = c2 * r2 * (global_best_position - particle['position'])
        particle['velocity'] = w * particle['velocity'] + cognitive_component + social_component
        particle['position'] = np.round(particle['position'] + particle['velocity'])

        fitness = evaluate_fitness(particle['position'])
        if fitness < particle['personal_best_fitness']:
            particle['personal_best_position'] = particle['position']
            particle['personal_best_fitness'] = fitness

        if fitness < global_best_fitness:
            global_best_position = particle['position']
            global_best_fitness = fitness

print("Best assignment of bicycles in stations:", global_best_position)
print("Remaining demand:", sum(demanda_estaciones) - sum(global_best_position))
print("Total bicycles assigned:", sum(global_best_position))
```
